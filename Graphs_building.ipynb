```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel('wiki.xlsx')

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.plot(df[df['cores']==core].sort_values('n_use')['n_use'],df[df['cores']==core].sort_values('n_use')['time'],label=str(core) + ' cores', marker='o')
plt.legend()
plt.grid(True)
plt.xlabel('n_use')
plt.ylabel('time')
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.semilogy(df[df['cores']==core].sort_values('n_use')['n_use'],df[df['cores']==core].sort_values('n_use')['heat rate'],label=str(core)+ ' cores', marker='o')
plt.legend()
plt.xlabel('n_use')
plt.ylabel('heat rate')
plt.grid(which='minor')
plt.grid(True)
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.semilogy(df[df['cores']==core].sort_values('n_use')['n_use'],df[df['cores']==core].sort_values('n_use')['score'],label=str(core)+ ' cores', marker='o')
plt.legend()
plt.xlabel('n_use')
plt.ylabel('score')
plt.grid(which='minor')
plt.grid(True)
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.scatter(df[df['cores']==core]['time'],df[df['cores']==core]['score'],label=str(core)+ ' cores')
plt.legend()
plt.xlabel('time')
plt.ylabel('score')
plt.grid(True)
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.scatter(df[df['cores']==core]['time'],df[df['cores']==core]['heat rate'],label=str(core)+ ' cores')
plt.legend()
plt.xlabel('time')
plt.ylabel('heat rate')
plt.grid(True)
plt.show()
```
