```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('wiki.xlsx')

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.plot(df[df['cores']==core].sort_values('n_use')['n_use'],df[df['cores']==core].sort_values('n_use')['time'],label=str(core) + ' cores', marker='o')
plt.legend()
plt.grid(True)
plt.xlabel('n_use')
plt.ylabel(r'time, $s$')
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.semilogy(df[(df['cores']==core) & (df['n_use']<=20000)].sort_values('n_use')['n_use'],1 + 1e-6 - df[(df['cores']==core) & (df['n_use']<=20000)].sort_values('n_use')['heat rate'],label=str(core)+ ' cores', marker='o')

plt.plot([0,20100], [0.01,0.01], '--', label=r'$0.99$ accuracy')
plt.legend()
plt.xlabel('n_use')
plt.ylabel('1 - heat rate')
plt.grid()
plt.grid(True)
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.semilogy(df[df['cores']==core].sort_values('n_use')['n_use'],1  - df[df['cores']==core].sort_values('n_use')['score'],label=str(core)+ ' cores', marker='o')
plt.legend()
plt.plot([0,700], [0.01,0.01], '--', label=r'$0.99$ accuracy')
plt.xlabel('n_use')
plt.plot([0,40100], [0.01,0.01], '--', label=r'$0.99$ accuracy')
plt.ylabel('1 - score')
plt.grid()
plt.grid(True)
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.semilogy(df[df['cores']==core]['time'],1 - df[df['cores']==core]['score'],label=str(core)+ ' cores', marker='o')
plt.legend()
plt.xlabel(r'time, $s$')
plt.ylabel('1 - score')
plt.grid(True)
plt.show()

plt.figure(figsize = (8,6))
plt.rcParams.update({'font.size': 12})
for core in set(df['cores']):
    plt.semilogy(df[(df['cores']==core) & (df['n_use']<=20000)]['time'],1 + 1e-6 - df[(df['cores']==core)  & (df['n_use']<=20000)]['heat rate'],label=str(core)+ ' cores', marker='o')
plt.legend()
plt.plot([0,400], [0.01,0.01], '--', label=r'$0.99$ accuracy')
plt.xlabel(r'time, $s$')
plt.ylabel('1 - heat rate')
plt.grid(True)
plt.show()
```
