```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import cv2
```

```python
words_GT_wiki = np.load('top words/words_WIKI_GT.txt',allow_pickle=True)
counts_GT_wiki = np.load('top words/counts_WIKI_GT.txt',allow_pickle=True)

words_GT = np.load('top words/words_GT.txt',allow_pickle=True)
counts_GT = np.load('top words/counts_GT.txt',allow_pickle=True)
```

```python
wiki_dict  = {}
for i, word in enumerate(words_GT_wiki):
    wiki_dict[word] = counts_GT_wiki[i]
```

```python
wp_dict  = {}
for i, word in enumerate(words_GT):
    wp_dict[word] = counts_GT[i]
```

### Wikipedia

```python
img = np.uint8(cv2.imread('top words/cloud.png'))
wordcloud = WordCloud(stopwords = {}, mask = img).generate_from_frequencies(wiki_dict)

f = {'size' : 15}
plt.rc('font', **f)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Top words in Wikipedia')
plt.axis('off')
plt.savefig('top words/Word Clouds/word_cloud_wiki1.png')
plt.show()
```

```python
img = np.uint8(cv2.imread('top words/wiki.png'))
wordcloud = WordCloud(stopwords = {}, mask = img).generate_from_frequencies(wiki_dict)

f = {'size' : 15}
plt.rc('font', **f)
plt.figure(figsize=(20,20))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Top words in Wikipedia')
plt.axis('off')
plt.savefig('top words/Word Clouds/word_cloud_wiki2.png')
plt.show()
```
